```python
#import dependencies
import pandas as pd
```

```python
#import data, read, and store data as a dataframe
file_path = "Data/melb_data.csv"
melb_data = pd.read_csv(file_path)
```

```python
melb_data
```

```python
melb_data.info()
```

```python
melb_data.describe()
```

```python
melb_data = melb_data.dropna(axis=0)
```

```python
melb_data.info()
```

```python
#prediction target
y = melb_data.Price
```

```python
#choosing feature, can use other methods 
#to determine number of features necessary 
features = ['Rooms', 'Bathroom', 'Landsize']
X = melb_data[features]
X.head()
```

```python
#Build Model
```

```python
from sklearn.tree import DecisionTreeRegressor

melb_model = DecisionTreeRegressor(random_state =1)
```

```python
#fit model
melb_model.fit(X,y)
```

```python
print(X.head())
print("Price Predictions are the following for the following",melb_model.predict(X.head()))
```
